```python
def read_file(file_name) :
    file = open(file_name, 'r+')
    f_list = file.readlines()
    file.close()
    f_list = [s.replace('\n', '') for s in f_list]
    return f_list

test_1 = read_file("test1.txt")
test_2 = read_file("test2.txt")
test_3 = read_file("test3.txt")
test_1
```

```python
class Obj:   ## 'h' = head, 't' = tail, 'b' = body, 'c' = candy with capacity
    def __init__(self, y, x, node_type, capacity = 0) : 
        self.x = x
        self.y = y
        self.type = node_type
        self.real_type = node_type
        self.capacity = capacity
    
    def eat_one(self):
        self.capacity = self.capacity - 1
    def __lt__(self, other):
        return False
```

```python
import copy
class Game:
    def __init__(self, table):
        self.map = self.set_map(table)
        self.snake = self.set_snake(table)
        self.current_state = self.get_init_state(table)
        self.current_stateCopy = copy.deepcopy(self.current_state)
        self.snake_copy = copy.deepcopy(self.snake)
#         self.qset = set([])
        
    def set_map(self, table):
        map_size = [int(s) for s in table[0].split(',')]
        return map_size
    
    def set_snake(self, table):
        head_coords = [int(s) for s in table[1].split(',')]
        return head_coords
    
    def get_init_state(self, table):
        current_state = []
        current_state0 = []
        for i in range(3, 3 + int(table[2])):
            candy_info = table[i].split(',')
            new_candy = Obj(int(candy_info[0]), int(candy_info[1]), 'c', int(candy_info[2]))
            current_state0.append(new_candy)
            
        current_hash = self.make_hash(current_state0)
        current_path = ''
        current_state.append(current_state0)
        current_state.append(current_hash)
        current_state.append(current_path)
#         current_state.append(self.snake)
        return current_state
    
    def make_hash(self, obj_list):
        current_hash = ''
        for i in self.snake:
            current_hash += str(i) + ','
        current_hash = current_hash[:-1]
        current_hash += '|'
        for i in obj_list:
            if (i.type == 'c'):
                current_hash += str(i.y) + ',' + str(i.x) + ',' + str(i.capacity) + '#'
        current_hash = current_hash[:-1] 
        return current_hash
    
    
    def get_candy_count(self):
        ans = 0
        for i in range(0, len(self.current_state[0])) :
            if (self.current_state[0][i].type == 'c'
                and self.current_state[0][i].capacity > 0):
                ans += 1
        return ans
    
    def get_object(self, y, x):
        if (len(self.snake) > 2):
            for i in range(2, len(self.snake) - 1):
                if (self.snake[i] == y and self.snake[i+1] == x):
                    if (i == (len(self.snake) - 2)):
                        return (0,'t')
                    else:
                        return (0,'b')
        for i in range(len(self.current_state[0])):
            if (self.current_state[0][i].x == x and self.current_state[0][i].y == y):
                return (i,'c')

        return (0,'e')
        
    
    def predict_snake(self, snake, direction):
        new_snake = []
        snake_head = snake[0:2]
        if (direction == 'u'):
            if (snake_head[0] == 0):
                snake_head[0] = self.map[1] - 1
            else:
                snake_head[0] = snake_head[0] - 1
        elif (direction == 'd'):
            if (snake_head[0] == (self.map[1] - 1)):
                snake_head[0] = 0
            else:
                snake_head[0] = snake_head[0] + 1
        elif (direction == 'r'):
            if (snake_head[1] == (self.map[0] - 1)):
                snake_head[1] = 0
            else:
                snake_head[1] = snake_head[1] + 1
        elif (direction == 'l'):
            if (snake_head[1] == 0):
                snake_head[1] = self.map[0] - 1
            else:
                snake_head[1] = snake_head[1] - 1
        
        for i in range(0, len(snake) - 2):
            new_snake.append(snake[i])
        new_snake.insert(0, snake_head[1])
        new_snake.insert(0, snake_head[0])
        return new_snake

    def special_predict(self, snake, direction):
        snake_head = snake[0:2]
        if (direction == 'u'):
            if (snake_head[0] == 0):
                snake_head[0] = self.map[1] - 1
            else:
                snake_head[0] = snake_head[0] - 1
        elif (direction == 'd'):
            if (snake_head[0] == (self.map[1] - 1)):
                snake_head[0] = 0
            else:
                snake_head[0] = snake_head[0] + 1
        elif (direction == 'r'):
            if (snake_head[1] == (self.map[0] - 1)):
                snake_head[1] = 0
            else:
                snake_head[1] = snake_head[1] + 1
        elif (direction == 'l'):
            if (snake_head[1] == 0):
                snake_head[1] = self.map[0] - 1
            else:
                snake_head[1] = snake_head[1] - 1
        snake.insert(0, snake_head[1])
        snake.insert(0, snake_head[0])
        return snake
    
    def check_before_move(self, snake, direction, test_name):
        if (len(snake) == 2):
            index, obj_type = self.get_object(snake[0], snake[1])
            if (obj_type == 'c'):
                if (self.current_state[0][index].capacity > 0):
                    
                    new_snake = self.special_predict(snake, direction)
                    
                    
                    new_game = Game(test_name)
                    new_game.snake = new_snake
                    
                    state0 = []
                    for i in self.current_state[0] :
                        obj = Obj(i.y, i.x, i.type, i.capacity)
                        state0.append(obj)
                        
                    new_game.current_state[0] = state0
                    new_game.current_state[0][index].eat_one()
                    new_game.current_state[1] = new_game.make_hash(new_game.current_state[0])
                    new_game.current_state[2] = new_game.current_state[2] + direction
                    return new_game
                    
            else:
                new_snake = self.predict_snake(snake, direction)
                
                new_game = Game(test_name)
                new_game.snake = new_snake
                
                state0 = []
                for i in self.current_state[0] :
                    obj = Obj(i.y, i.x, i.type, i.capacity)
                    state0.append(obj)

                    
                new_game.current_state[0] = state0
                new_game.current_state[1] = new_game.make_hash(new_game.current_state[0])
                new_game.current_state[2] = new_game.current_state[2] + direction
                return new_game
            

        else:
            curr_index, curr_obj_type = self.get_object(snake[0], snake[1])
            if (curr_obj_type == 'c' and self.current_state[0][curr_index].capacity > 0):
                new_snake = self.special_predict(snake, direction)
                next_index, next_obj_type = self.get_object(new_snake[0], new_snake[1])
                if (next_obj_type == 'b' or next_obj_type == 't'):
                    return None
                else:
                    
                    new_game = Game(test_name)
                    
                    state0 = []
                    for i in self.current_state[0] :
                        obj = Obj(i.y, i.x, i.type, i.capacity)
                        state0.append(obj)
                        
                    new_game.current_state[0] = state0
                    new_game.current_state[0][curr_index].eat_one()
                    new_game.snake = new_snake
                    new_game.current_state[1] = new_game.make_hash(new_game.current_state[0])
                    new_game.current_state[2] = new_game.current_state[2] + direction
                    
                    return new_game
                
            else:
                new_snake = self.predict_snake(snake, direction)
                next_index, next_obj_type = self.get_object(new_snake[0], new_snake[1])
                if (next_obj_type == 'b' or next_obj_type == 't'):
                    return None
                else:
                    new_game = Game(test_name)
                    new_game.snake = new_snake
                    
                    state0 = []
                    for i in self.current_state[0] :
                        obj = Obj(i.y, i.x, i.type, i.capacity)
                        state0.append(obj)
                        
                    new_game.current_state[0] = state0
                    new_game.current_state[1] = new_game.make_hash(new_game.current_state[0])
                    new_game.current_state[2] = new_game.current_state[2] + direction
                    return new_game
                    
    def is_repetitive_state(self) : 
        return (self.current_state[1] in self.qset)             
    
    def get_copy(self, state): 
        copy = []
        state0 = []
        for i in state[0] :
            obj = Obj(i.y, i.x, i.type, i.capacity)
            state0.append(obj)
        state1 = state[1]
        state2 = state[2]
        copy.append(state0)
        copy.append(state1)
        copy.append(state2)
        return copy
    
    
    def last_round(self):
        if (self.get_candy_count() == 1):
            y = game.snake[0]
            x = game.snake[1]
            for i in self.current_state[0]:
                if (i.x == x and i.y == y):
                    return True
    
    
game = Game(test_1)        
# print(game.get_object(3,1))
print(game.special_predict([0,0,1,0], 'd'))
```

<h1 style = "text-align : center">BFS (Breathe First Search)</h1>
<div style = "align-items: center">
    <img src = "img3.png">
</div>
<h2>Algorithm:</h2>
<p>First of all we will start from the <b>root node : (initial state)</b>, then per every node in graph (in our problem: state), <b>we will visit all of its children</b>, and per every feasible child we will append its state in a q named <b>queue</b>, then for next iterations, we pop from head of this queue and visit its children again and so on.</p>
<p>We will finish iteration in these ciscumstances: </p>
<p style="text-indent :2em;">1. Number of Candies with positive capacity equals to 0 which means we have found the correct path</p>
<p style="text-indent :2em;">2. Length of queue equals to 0 which means there are no valid path.</p>
<br>
<p>For every state we have 4 possibility: up, down, right, and left. and to check feasibility of these possibilities we use two functions: </p>
<p style="text-indent :2em;">1. check_before_move() which checkes whether new state is possible or not. </p>
<p style="text-indent :2em;">2. is_repetitive_state() which checkes whether this state is repetitive or not. <b>Checking is done with hash strings</b> (we store every visited state in a set to reduce time consumption of checking).</p>
<br>
<p>if condition 1 is correct: we should restore previous state from copy_state and increament <b>states</b> which holds total states</p>
<p>If everything is good, it means this state is feasible and we add this state into queue and increament <b>u_states</b> which holds unique states</p>
<p>Ultimately if we find a path, we will return: <b>path, states, and u_states</b></p>
<br>
<h2>Advantages : </h2>
<br>
<p>Completeness : because we check every possible state.</p>
<p>Optimality : because we first check pathes with shorter length (1,2,3,...) so it will result optimality.</p>
<h2>Disadvantages: </h2>
<p>Number of states : we check every possible state in a certain length of path.</p>
<p>Complexity of space : because we must save every state we have seen which means all possible states.</p>



```python
def bfs(map_index):
    
    if (map_index == 1) :
        file_name = test_1
    elif (map_index == 2) :
        file_name = test_2
    elif (map_index == 3) :
        file_name = test_3

    game = Game(file_name)    
    
    queue = []
    sec_queue = []
    
    visited = set()
    hash_codes = set()
    
    queue.append(game)
    hash_codes.add(game.current_state[1])
    
    unq_states = 1
    states = 1
    init = True
    finished = False
    
    moves = ['l', 'd', 'u', 'r']
    while len(queue):
        
        new_game = queue.pop(0)
        hash_codes.remove(new_game.current_state[1])
#         print(new_game.current_state[1] in hash_codes)
        visited.add(new_game.current_state[1])
        
        if (new_game.get_candy_count() == 0):
            return [new_game.current_state[2], unq_states, states]
        
        for i in moves:
            new_child = new_game.check_before_move(new_game.snake, i, file_name)
            if(new_child != None):
                if (new_child.current_state[1] in visited):
                    states += 1
                elif (new_child.current_state[1] in hash_codes):
                    pass
                else:
                    queue.append(new_child)
#                     visited.add(new_child.current_state[1])
                    hash_codes.add(new_child.current_state[1])
                    unq_states += 1
                    states += 1
```

```python
import time
def bfs(i) : 
    if (i == 1) : 
        game = Game(test_1)
    elif (i == 2) : 
        game = Game(test_2)
    elif (i == 3) : 
        game = Game(test_3)
        
    start = time.time()
    response = game.bfs()
    end = time.time()
    
    duration = end - start
    path = response[0]
    uStates = response[1]
    states = response[2]
    
    return [path, states, uStates, duration]

bfs(1)
```
