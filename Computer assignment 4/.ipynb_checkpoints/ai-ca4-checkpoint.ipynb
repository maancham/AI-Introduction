### Numeric columns missing val handling:
<p style="text-indent :2em;">1. <mark>Constant value</mark>: Replace it with a constant value. Typically, this is used in discussion with the domain expert for the data we are dealing with.</p> 
<p style="text-indent :2em;">2. <mark>Mean or median</mark>:Replace it with the mean or the median. This is a decent approach especially when the data size small, but it does add bias.</p> 
<p style="text-indent :2em;">3. <mark>Based on other columns</mark>:Replace it with values by making use of information from other columns.</p>

### Categorical columns missing val handling:
<p style="text-indent :2em;">1. <mark>Constant value</mark>: Replace it with a constant value or the most popular category. This is a good approach especially when the data size is small — but it does add <b>bias</b>.</p> 
<p style="text-indent :2em;">2. <mark>Classification</mark>:One can also create a classification model.By using prediction methods we can fill in the missing values based on information and processing on other columns. </p> 
<p style="text-indent :2em;">3. <mark>Unknown</mark>:The most common and popular approach is to model the missing value in a categorical column as a new category called “Unknown.” <b> We use this method for most our of categorical columns</b> </p>


## -- Fourth section:No, each one has to be handled in its own way.
<p style="text-indent :2em;">1. <mark>Binary Feature Encoding</mark>: Binary features are those with only two possible values. we can use Panda’s replace() to encode them.</p>
<p style="text-indent :2em;">2. <mark>Ordinal Feature Encoding</mark>: Ordinal features are those with some order associated with them. We can tell from the sample of ordinal features below these features have an order that may be important.For ordinal features, we use integer encoding. To integer encode our data we simply convert labels to integer values.We can label encode data with <b>Sklearn’s LabelEncoder()</b>. While using LabelEncoder() is very quick and easy, <mark>it may not be the best choice here</mark>: the order of our encoding is not exactly right. Also, we had to handle our null values before being able to use it.<b> Panda’s map()</b> substitutes each value with another specified value, similar to replace() that we used above. We have four types of ordinal columns. Each type has been handled separately.</p> 
<p style="text-indent :2em;">3. <mark>Nominal Features</mark>: Nominal features are categorical features that have no numerical importance. Order does not matter.<b>One-hot encoding</b> is a better technique when order doesn’t matter.Despite the different names, the basic strategy is to convert each category value into a new column and assigns a 1 or 0 (True/False) value to the column. This has the benefit of not weighting a value improperly but does have the downside of adding more columns to the data set.</p>

```python
def ordinal_encode():
    ord1_dict = {"nonexist":0, "Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}
    ordinal_columns1 = [
        'ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'HeatingQC',
        'KitchenQual', 'FireplaceQu', 'GarageQual', 'GarageCond', 'PoolQC']
    for i in ordinal_columns1:
        train_df[i] = train_df[i].map(ord1_dict)

    ord2_dict = {"N":0, "Y":1}
    ordinal_columns2 = ['CentralAir']
    for i in ordinal_columns2:
        train_df[i] = train_df[i].map(ord2_dict)

    ord3_dict = {"AllPub":4, "NoSewr":3, "NoSeWa":2, "ELO":1}
    ordinal_columns3 = ['Utilities']
    for i in ordinal_columns3:
        train_df[i] = train_df[i].map(ord3_dict)

    ord4_dict = {"nonexist":0, "Unf":1, "LwQ":2, "Rec":3, "BLQ":4, "ALQ":5, "GLQ":6}
    ordinal_columns4 = ['BsmtFinType1', 'BsmtFinType2']
    for i in ordinal_columns4:
        train_df[i] = train_df[i].map(ord4_dict)
    
    ord5_dict = {"nonexist":0, "No":1, "Mn":2, "Av":3, "Gd":4}
    ordinal_columns5 = ['BsmtExposure']
    for i in ordinal_columns5:
        train_df[i] = train_df[i].map(ord5_dict)
        
               
nominal_columns = ['Foundation', 'MSZoning', 'Alley', 'LotShape','LandContour', 'Street',
                   'LotConfig', 'Neighborhood', 'HouseStyle', 'RoofMatl', 'BldgType', 'LandSlope',
                    'Condition1', 'RoofStyle', 'Condition2', 'Exterior1st', 'Exterior2nd', 'MasVnrType',
                   'Heating', 'Electrical', 'Functional', 'GarageType', 'SaleType', 'SaleCondition',
                  'GarageFinish', 'PavedDrive', 'MiscFeature', 'Fence']

train_df = pd.get_dummies(train_df, columns = nominal_columns, drop_first = True)

    
ordinal_encode()
train_df
```

### linear regression:
<p> Linear regression is a statistical model that examines the linear relationship between two (Simple Linear Regression ) or more (Multiple Linear Regression) variables — a dependent variable and independent variable(s). Linear relationship basically means that when one (or more) independent variables increases (or decreases), the dependent variable increases (or decreases) too. There are two main ways to perform linear regression in Python — with <mark>Statsmodels and scikit-learn.</mark> Explanation of some used methods: </p>
<p style="text-indent :2em;"><b>train_test_split:</b> Split arrays or matrices into random train and test subsets. Test size is set to be 0.2 of total dataframe. Random state controls the shuffling applied to the data before applying the split. Pass an int for reproducible output across multiple function calls.For optimum situation and <mark>since we only have one parameter to optimize,</mark> A loop is used to find best random_state number.</p>
<p style="text-indent :2em;"><b>reshape:</b> Gives a new shape to an array without changing its data. Needed to make a 2d vector out of y_train, y_test and predictions.</p>

#### As you can see at the end of this block, there are 63 random_state values that are qualified.

```python
mae_results = []
rmse_results = []
x_value = [x for x in range(500)]
for i in range(500):
    X_train, X_test, y_train, y_test = train_test_split(train_df.drop('SalePrice', axis=1), 
                                                        train_df['SalePrice'], test_size=0.2, random_state=i)
    y_train= y_train.values.reshape(-1,1)
    y_test= y_test.values.reshape(-1,1)
    lm = LinearRegression()
    lm.fit(X_train,y_train)
    predictions = lm.predict(X_test)
    predictions= predictions.reshape(-1,1)
    mae_results.append(metrics.mean_absolute_error(y_test, predictions))
    rmse_results.append(np.sqrt(metrics.mean_squared_error(y_test, predictions)))
    
plt.figure(figsize=(20, 10))
plt.plot(x_value,mae_results, c='blue')
plt.plot(x_value,rmse_results, c='red')
plt.show()
linear_df = pd.DataFrame({'mae': mae_results, 'rmse': rmse_results, 'random_value': x_value})
linear_df[(linear_df['mae'] <= 20000) & (linear_df['rmse'] <= 30000)].sort_values(by=['rmse'])
```

### Knn:
<p> k-Nearest-Neighbors (k-NN) is a supervised machine learning model. Supervised learning is when a model learns from data that is already labeled. k-NN models work by taking a data point and looking at the ‘k’ closest labeled data points. The data point is then assigned the label of the majority of the ‘k’ closest points. For example, if k = 5, and 3 of points are ‘green’ and 2 are ‘red’, then the data point in question would be labeled ‘green’, since ‘green’ is the majority. This makes training faster and testing phase slower and costlier. Costly testing phase means time and memory. In the worst case, KNN needs more time to scan all data points and scanning all data points will require more memory for storing training data. </mark> Explanation of some used methods: </p>
<p style="text-indent :2em;"><b>train_test_split:</b> Split arrays or matrices into random train and test subsets. Test size is set to be 0.2 of total dataframe. Random state controls the shuffling applied to the data before applying the split. Pass an int for reproducible output across multiple function calls.For optimum situation and <mark>since we only have one parameter to optimize,</mark> A loop is used to find best random_state number.</p>

#### Note that there are two parameters to consider here. First one being random_state and second being n_neighbors. For each n_neighbor value, 100 random_state values has been analysed and their best(min mae/rmse) is chosen to represent result of that n_neighbor.

```python
mae_results = []
rmse_results = []
min_indexs = []
x_value = [x for x in range(29)]
for i in range(1,30):
    mae_temp = []
    rmse_temp = []
    for j in range(100):
        X_train, X_test, y_train, y_test = train_test_split(train_df.drop('SalePrice', axis=1), 
                                                        train_df['SalePrice'], test_size=0.2, random_state=j)
        knn = KNeighborsRegressor(n_neighbors=i)
        knn.fit(X_train, y_train)
        predictions = knn.predict(X_test)
        mae_temp.append(metrics.mean_absolute_error(y_test, predictions))
        rmse_temp.append(np.sqrt(metrics.mean_squared_error(y_test, predictions)))

    min_index = mae_temp.index(min(mae_temp))
    min_indexs.append(min_index)
    mae_results.append(mae_temp[min_index])
    rmse_results.append(rmse_temp[min_index])
    
    
plt.figure(figsize=(20, 10))
plt.plot(x_value,mae_results, c='blue')
plt.plot(x_value,rmse_results, c='red')
plt.show()

super_min_idx = mae_results.index(min(mae_results))
print("Best Results:")
print("MAE: ", mae_results[super_min_idx])
print("RMSE: ", rmse_results[super_min_idx])
print("N_neighbors: ", super_min_idx)
print("Random_state: ", min_indexs[super_min_idx])
```

### Decison tree:
<p> Decision Tree is one of the popular and most widely used Machine Learning Algorithms because of its robustness to noise, tolerance against missing information, handling of irrelevant, redundant predictive attribute values, low computational cost, interpretability, fast run time and robust predictors. </mark> Explanation of some used methods: </p>
<p style="text-indent :2em;"><b>train_test_split:</b> Split arrays or matrices into random train and test subsets. Test size is set to be 0.2 of total dataframe. Random state controls the shuffling applied to the data before applying the split. Pass an int for reproducible output across multiple function calls.For optimum situation and <mark>since we only have one parameter to optimize,</mark> A loop is used to find best random_state number.</p>
<p style="text-indent :2em;"><b>max_depth:</b> The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples. The theoretical maximum depth a decision tree can achieve is one less than the number of training samples, but no algorithm will let you reach this point for obvious reasons, one big reason being overfitting.</p>

#### Note that there are two parameters to consider here. First one being random_state and second being max_depth. For each max_depth value, 50 random_state values has been analysed and their best(min mae/rmse) is chosen to represent result of that max_depth. Note that I've tried values under 15 for max_depth and their results were not acceptable so values above 15 have been tested.

```python
mae_results = []
rmse_results = []
min_indexs = []
x_value = [x for x in range(15,35)]
for i in range(15,35):
    for j in range(20,70):
        mae_temp = []
        rmse_temp = []
        X_train, X_test, y_train, y_test = train_test_split(train_df.drop('SalePrice', axis=1), 
                                                                train_df['SalePrice'], test_size=0.2, random_state=j)
        classifier = DecisionTreeRegressor(max_depth= i)
        classifier.fit(X_train, y_train)
        predictions = classifier.predict(X_test)
        mae_temp.append(metrics.mean_absolute_error(y_test, predictions))
        rmse_temp.append(np.sqrt(metrics.mean_squared_error(y_test, predictions)))
        
    min_index = mae_temp.index(min(mae_temp))
    min_indexs.append(min_index)
    mae_results.append(mae_temp[min_index])
    rmse_results.append(rmse_temp[min_index])

plt.figure(figsize=(20, 10))
plt.plot(x_value,mae_results, c='blue')
plt.plot(x_value,rmse_results, c='red')
plt.xlabel("max_depth")
plt.show()
super_min_idx = mae_results.index(min(mae_results))
print("Best Results:")
print("MAE: ", mae_results[super_min_idx])
print("RMSE: ", rmse_results[super_min_idx])
print("Max_depth: ", super_min_idx + 15)
print("Random_state: ", min_indexs[super_min_idx] + 20)
```

## -- Third section:

<p> <b>Overfitting</b> is the case where the overall cost is really small, but the generalization of the model is unreliable. This is due to the model learning “too much” from the training data set.</p>
<p> <b>Underfitting</b> is the case where the model has “ not learned enough” from the training data, resulting in low generalization and unreliable predictions.</p>

<h4> <mark>Overfitting</mark> is when the model’s error on the training set (i.e. during training) is very low but then, the model’s error on the test set (i.e. unseen samples) is large!. Underfitting is when the model’s error on both the training and test sets (i.e. during training and testing) is very high. </h4>

### Decision tree:

# Phase 3:
## -- First section:
<p> A random forest is a meta estimator that fits a number of classifying decision trees on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting. </p>
<p style="text-indent :2em;">1. <mark>n_estimators</mark>: The number of trees in the forest. For this parameter, 10 values have been cosidered and tested.</p> 
<p style="text-indent :2em;">2. <mark>max_depth</mark>: The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples. Values ranging form 10 to 30 have been tested for this parameter.</p> 

```python
from sklearn.ensemble import RandomForestRegressor
mae_results = []
rmse_results = []
min_indexs = []
X_train, X_test, y_train, y_test = train_test_split(train_df.drop('SalePrice', axis=1), 
                                                        train_df['SalePrice'], test_size=0.2, random_state=42)
depth_list = list(range(10,31))
est_list = [10,30,50,70,80,100]
for dp in depth_list:
    mae_temp = []
    rmse_temp = []
    for est in est_list:
        rf = RandomForestRegressor(n_estimators = est, max_depth = dp)
        rf.fit(X_train, y_train)
        predictions = rf.predict(X_test)
        mae_temp.append(metrics.mean_absolute_error(y_test, predictions))
        rmse_temp.append(np.sqrt(metrics.mean_squared_error(y_test, predictions)))
        
    min_index = mae_temp.index(min(mae_temp))
    min_indexs.append(min_index)
    mae_results.append(mae_temp[min_index])
    rmse_results.append(rmse_temp[min_index])
    
plt.figure(figsize=(20, 10))
plt.plot(depth_list,mae_results, c='blue')
plt.plot(depth_list,rmse_results, c='red')
plt.xlabel('max_depth')
plt.show()
super_min_idx = rmse_results.index(min(rmse_results))
print("Best Results:")
print("MAE: ", mae_results[super_min_idx])
print("RMSE: ", rmse_results[super_min_idx])
print("Max_depth: ", super_min_idx + 10)
print("N_estomators: ", est_list[min_indexs[super_min_idx]])
```

## -- Second section:
<p> A voting regressor is an ensemble meta-estimator that fits several base regressors, each on the whole dataset. Then it averages the individual predictions to form a final prediction. </p>
<p style="text-indent :2em;">1. <mark>estimators</mark>: Invoking the fit method on the VotingRegressor will fit clones of those original estimators that will be stored in the class attribute self.estimators_.</p> 
<p style="text-indent :2em;">2. <mark>weights</mark>: Sequence of weights (float or int) to weight the occurrences of predicted values before averaging. Uses uniform weights if None.</p>

```python
from sklearn.ensemble import VotingRegressor
X_train, X_test, y_train, y_test = train_test_split(train_df.drop('SalePrice', axis=1), 
                                                        train_df['SalePrice'], test_size=0.2, random_state=50)
min_mae = 100000
min_rmse = 100000
for j in range(50):
    lm = LinearRegression()
    lm.fit(X_train,y_train)
    knn = KNeighborsRegressor(n_neighbors=19)
    knn.fit(X_train, y_train)
    classifier = DecisionTreeRegressor(max_depth= 30)
    classifier.fit(X_train, y_train)

    voting_reg = VotingRegressor(estimators=[('Knn', knn), ('DTree', classifier), ('LogReg', lm)], weights = [2,5,3])
    voting_reg.fit(X_train,y_train)
    predictions = voting_reg.predict(X_test)
    mae = metrics.mean_absolute_error(y_test, predictions)
    if (mae < min_mae):
        min_mae = mae
        min_rmse = np.sqrt(metrics.mean_squared_error(y_test, predictions))

print("MAE: ", min_mae)
print("RMSE: ", min_rmse)
```

## -- Third section:
<p> There are some linear regression models that have better results than voring regression. However, voting regression is better than all other methods. The reason is that you can take the best parameter-tuned version of each of the models(in our example linear, decision tree and knn). Also, you have the chance to rank each one of the used models by using weight parameter to give weight to some models(since their solo results were better than others). As we can see above, voting regression has an advantage over knn and decision tree and a better generality over linear regression. </p>
